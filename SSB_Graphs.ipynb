```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set() #Set style
sns.set_style('ticks',{'font.family':'serif', 'font.serif':'Times New Roman'})
sns.set_context('paper', font_scale=1.8)
sns.set_palette(sns.color_palette("Paired"))
```

# 1 real field case

```python
phi = np.arange(-40,40,0.1)
v = 0.5 * phi**2 + 0.01/24 * phi**4

d = {'$\phi$': phi, '$V(\phi)$': v }
df = pd.DataFrame(data = d)
```

```python
plt.figure(figsize = [9.2,4.9], dpi = 300)

plt.axhline(y=0, color='k', lw = 1, linestyle='--')
plt.plot(phi,v, lw = 2, color = 'royalblue', label = r'$V(\phi) = \frac{}{}$')
#plt.axhline(y=0, color='k', linestyle='--')
plt.xlabel(r'$\phi$')
plt.ylabel(r'$V(\phi)$')
plt.xticks([0])
plt.yticks([0])
plt.savefig('SSB_Graphs/PlusPotential_1D.pdf', dpi =300)
plt.show()
```

```python
v_SSB = 0.01/24 * phi**4 - 0.5 * phi**2
```

```python
plt.figure(figsize = [9.2,4.9], dpi = 300)

plt.axhline(y=0, color='k', lw = 1, linestyle='--')
plt.plot(phi,v_SSB, lw = 2, color = 'royalblue')
plt.xlabel(r'$\phi$')
plt.ylabel(r'$V_{SMB}(\phi)$')
plt.xticks([0])
plt.yticks([0])

plt.savefig('SSB_Graphs/MinusPotential_1D.pdf', dpi =300)
plt.show()
```

# 2 real field case

```python
x = y = np.arange(-30,30,0.1)

x, y = np.meshgrid(x, y)


z = np.sqrt(x**2 + y**2)

phi_1 = phi_2 = np.arange(-30,30,0.1)

phi_1, phi_2 = np.meshgrid(phi_1, phi_2)


v_3D = 0.5 * phi_1**2 + 0.5 * phi_2**2 + 0.01/24 * (phi_1**2 + phi_2**2)**2

v_3D_SSB = -0.5 * phi_1**2 - 0.5 * phi_2**2 + 0.01/24 * (phi_1**2 + phi_2**2)**2
```

```python
fig = plt.figure(figsize = [9.2,4.9], dpi = 300)
ax = fig.add_subplot(projection='3d')

ax.plot_surface(phi_1, phi_2, v_3D, color = 'deepskyblue')
ax.set_ylabel(r'$\phi_2$')
ax.set_xlabel(r'$\phi_1$')
ax.set_zlabel(r'$V(\phi_1,\phi_2)$')
ax.set_zticks([0])
ax.set_xticks([0])
ax.set_yticks([0])

plt.savefig('SSB_Graphs/PlusPotential_2D.pdf', dpi =300)
plt.show()
```

```python
fig = plt.figure(figsize = [9.2,4.9], dpi = 300)
ax = fig.add_subplot(projection='3d')

ax.plot_surface(phi_1, phi_2, v_3D_SSB, color = 'deepskyblue')
ax.set_ylabel(r'$\phi_2$')
ax.set_xlabel(r'$\phi_1$')
ax.set_zlabel(r'$V_{SSB}(\phi_1,\phi_2)$')
ax.set_zticks([0])
ax.set_xticks([0])
ax.set_yticks([0])
#ax.set_zlim([-1,1])

plt.savefig('SSB_Graphs/MinusPotential_2D.pdf', dpi =300)
plt.show()
```
