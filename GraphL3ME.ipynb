```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xml.etree.ElementTree as ET
import os,glob,sys
import scipy as sc

sns.set() #Set style
sns.set_style('ticks',{'font.family':'serif', 'font.serif':'Times New Roman'})
sns.set_context('paper', font_scale=1.8)
sns.set_palette(sns.color_palette("Paired"))
```

```python
#variables
e = 1
k = 0.1 #1.83e-23
n = 1 #6e23

u = np.arange(-10,10,0.001)
```

```python
#equations

t = 1/((0.5*k/e)*np.log((e*n-u)/(u+e*n)))


#s = n*k*np.log(2*np.cosh(e/(k*t)))-(n*e/t)*np.tanh(e/(k*t)) 

s = n*k*(((-n*e-u)/(2*n*e))*np.log((n*e+u)/(2*n*e))-((n*e-u)/(2*n*e))*np.log((n*e-u)/(2*n*e)))
```

```python
#Plot
fig = plt.figure(figsize = [9.2,4.9], dpi = 300)

plt.plot(u,s, color = 'red')


#plt.ylim([-10,10])
plt.show()
```

```python
#Plot
fig = plt.figure(figsize = [9.2,4.9], dpi = 300)

plt.plot(u,t, color = 'indigo', lw = 2, label = r'T(U)')


plt.ylabel('Temperatura')
plt.xlabel('Energia interna')
plt.ylim([-1000,1000])
plt.xticks([0])
plt.yticks([0])
plt.legend()
plt.savefig('TxUGraph.pdf', dpi = 300)
plt.savefig('TxUGraph', dpi = 300)
plt.show()
```

```python
#Plot
fig = plt.figure(figsize = [9.2,4.9], dpi = 300)


plt.plot(u,s, color = 'red', label = 'S(U)')


plt.ylabel('Temperatura')
plt.xlabel('Energia interna')
#plt.ylim([-1000,1000])
plt.xticks([0])
plt.yticks([0])
plt.legend()
plt.savefig('SxUGraph.pdf', dpi = 300)
plt.savefig('SxUGraph', dpi = 300)
plt.show()
```
