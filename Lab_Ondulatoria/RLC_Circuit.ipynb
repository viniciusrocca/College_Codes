```python
fig = plt.figure(figsize = [7,6])


#Fasor da tens√£o no capacitor
plt.arrow(0,0,9*np.cos(np.pi/6), 9*np.sin(np.pi/6), width = 0.025, head_width = 0.3, 
          color = 'royalblue', lw = 1.0 )
plt.annotate(r'$V_R$', xy=(9*np.cos(np.pi/6)/2, 9*np.sin(np.pi/6)/2), xytext=(10, -10),
             textcoords='offset points', color  = "royalblue")

#Fasor da corrente
plt.arrow(0,0,4*np.cos(np.pi/6), 4*np.sin(np.pi/6), width = 0.025, head_width = 0.3, 
          color = 'red', lw = 1.0)
plt.annotate(r'$I_m$', xy=(4*np.cos(np.pi/6)/2, 4*np.sin(np.pi/6)/2), xytext=(10, -10),
             textcoords='offset points', color  = "red")


#Fasor do indutor
plt.arrow(0,0,6*np.cos(2*np.pi/3), 6*np.sin(2*np.pi/3), width = 0.025, head_width = 0.3, 
          color = 'indigo', lw = 1.0 )
plt.annotate(r'$V_L$', xy=(6*np.cos(2*np.pi/3)/2, 6*np.sin(2*np.pi/3)/2), xytext=(10, -10),
             textcoords='offset points', color  = "indigo")

#Fasor do capacitor
plt.arrow(0,0,-2*np.cos(2*np.pi/3), -2*np.sin(2*np.pi/3), width = 0.025, head_width = 0.3, 
          color = 'darkgreen', lw = 1.0 )
plt.annotate(r'$V_C$', xy=(-2*np.cos(2*np.pi/3)/2, -2*np.sin(2*np.pi/3)/2), xytext=(10, -10),
             textcoords='offset points', color  = "darkgreen")

#Fasor capacitor - indutor
plt.arrow(0,0,4*np.cos(2*np.pi/3), 4*np.sin(2*np.pi/3), width = 0.025, head_width = 0.3, 
          color = 'magenta', lw = 1.0 )
plt.annotate(r'$V_L - V_C$', xy=(6*np.cos(2*np.pi/3)/2-3, 6*np.sin(2*np.pi/3)/2-1), xytext=(10, -10),
             textcoords='offset points', color  = "magenta")


plt.arrow(-10,0,18,0, width = 0.025, head_width = 0.3, color = 'black', lw = 1.0 )
plt.arrow(0,-5,0,13, width = 0.025, head_width = 0.3, color = 'black', lw = 1.0 )
plt.xlim(-9,10)
plt.ylim(-9,10)
plt.xticks([])
plt.yticks([])
plt.axis('off')
#plt.legend()
plt.savefig('Graphs/RLC/Phasor_diagram.pdf', dpi = 300)
plt.show()
```

## Dados coletados

```python
#Frequ√™ncia
f = np.arange(5,15.5, 0.5)

#Tens√£o
v = np.array([11.1,12.6,14.1,15.4,16.1,16.4,16.0,15.4,14.6,13.6,12.7,12.0,11.2,10.4, 9.86,9.20,8.72,8.24,7.84,7.52,7.20])
#Incerteza da tens√£o (3% do valor lido):
sigma_v = v*0.03
sigma_v = [float('{:g}'.format(float('{:.1g}'.format(sigma)))) for sigma in sigma_v]
```

```python
plt.figure(figsize = [9.2,4.9], dpi = 300) 
plt.errorbar(f,v,sigma_v,0,'o', ms = 4.5, capsize=3, color = 'royalblue', label = 'Dados')
plt.axvspan(7.5-0.3, 7.5+0.3 , color='lightsteelblue', alpha=0.3, lw=0, label = "Regi√£o de Resson√¢ncia")

plt.ylabel('Tens√£o (V)')
plt.xlabel('Frequ√™ncia (kHz)')
plt.xticks(np.arange(5,16,1))
plt.legend()
plt.savefig('Graphs/RLC/Dados_Vxf.pdf', dpi = 300)
plt.show()
```
