```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
W = 5.9e-6/4
mue = 9.284764e-24/1.602e-19
B = np.arange(0,0.5,1e-3)
E1 = W+B*mue
E2 = W-B*mue
E3 = -W + np.sqrt(4*W**2+ B**2 * mue**2)
E4 = -W - np.sqrt(4*W**2+ B**2 * mue**2)
```

```python

fig = plt.figure(figsize = [9.2,4.9], dpi = 300)
plt.plot(B,E1, linewidth = 1.5, color = 'darkviolet', label = r'$E = W+B\mu_e$')
plt.plot(B,E2, linewidth = 1.5, color = 'darkgreen',label = r'$E = W-B\mu_e$')
plt.plot(B,E3, linewidth = 1.5, color = 'dodgerblue', label = r'$E = -W+\sqrt{4W^2 + B^2 \mu_e^2}$')
plt.plot(B,E4, linewidth = 1.5, color = 'limegreen',label = r'$E = -W-\sqrt{4W^2 + B^2 \mu_e^2}$')
plt.legend()
plt.xlabel(r'Campo magnético [T]')
plt.ylabel(r'Energia [eV]')
plt.title(r'Autovalores de $\hat{H}$ em função de B')
plt.savefig('graph_3c')
plt.show()
```
