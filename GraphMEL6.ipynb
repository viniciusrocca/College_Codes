```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xml.etree.ElementTree as ET
import os,glob,sys
import scipy as sc

sns.set() #Set style
sns.set_style('ticks',{'font.family':'serif', 'font.serif':'Times New Roman'})
sns.set_context('paper', font_scale=1.8)
sns.set_palette(sns.color_palette("Paired"))
```

```python
xf = np.arange(-50,50,0.01)
xb = np.arange(-5,5,0.001)
y_f = 1/(np.exp(xf)+1)
y_b = 1/(np.exp(xb)-1)
```

```python
fig = plt.figure(figsize = [9.2,4.9], dpi = 300)

plt.plot(xf,y_f, color = 'indigo', lw = 2)


plt.ylabel(r'$<n_i>$')
plt.xlabel(r'$\frac{(\varepsilon_i-\mu)}{KT}$')
plt.title('Nmero mdio de ocupao de cada nvel para frmions')
plt.savefig('OccupationMeanValueF', dpi = 300)
plt.show()
```

```python
fig = plt.figure(figsize = [9.2,4.9], dpi = 300)

plt.plot(xb,y_b, color = 'royalblue', lw = 2)

plt.ylim([-10,10])
plt.ylabel(r'$<n_i>$')
plt.xlabel(r'$\frac{(\varepsilon_i-\mu)}{KT}$')
plt.title('Nmero mdio de ocupao de cada nvel para bsons')
plt.savefig('OccupationMeanValueB', dpi = 300)
#plt.yscale('log')
plt.show()
```
